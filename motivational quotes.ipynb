```python
import pandas as pd
def quotes(soup1):
    author_name = [i.text.strip() for i in soup1.find_all('span',class_='authorOrTitle')]
    quotes=[list(i)[0].strip() for i in soup1.find_all('div',class_='quoteText')]
    likes=[int(list(i)[0].strip().split()[0]) for i in soup1.find_all('a',class_='smallText')]
    tags=[i.text.split()[1:] for i in soup1.find_all('div',class_='greyText smallText left')]
    return pd.DataFrame({"Authors":author_name,"Quotes":quotes,"likes":likes,"tags":tags})
```

```python
df=quotes(soup1)
df
```

```python
df.shape
```

```python
df.to_csv("motivationalquotes.csv")
```

```python
studs= 'https://www.goodreads.com/quotes/tag/motivational?page={}'
```

```python
studs.format(3)
```

```python
tables=[]
from tqdm import tqdm
for i in tqdm(range(1,51)):
    url_all = studs.format(i)
    html_all = requests.get(url_all).text
    soup = BeautifulSoup(html_all,'html.parser')
    tables.append(quotes(soup))
```

```python
tables
```

```python
data=pd.concat(tables,axis=0)
```
